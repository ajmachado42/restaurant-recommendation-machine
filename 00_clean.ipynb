# Clean column names and dtypes

```python
# lower case all column names
col_new = []
col_old = df.columns
for col in col_old:
    col_new.append(col.lower())

# create old/new name dictionary
col_rename = dict(zip(col_old,col_new))

#rename in df
df = df.rename(columns=col_rename)
df.head()
```

```python
# drop column 7514
df = df.drop(columns = '7514')
```

```python
df.isnull().sum()
```

## Rating Dtype: Object to Float

```python
df['rating'].unique()
```

```python
# take care of 'Like' values

df[(df['rating'] == 'Like')]
df = df.drop(df[(df['rating'] == 'Like')].index)
df['rating'].unique()
```

```python
# take care of nan values

df['rating'] = df['rating'].fillna('None')
df = df.drop(df[(df['rating'] == 'None')].index)
df['rating'].unique()
```

```python
# change rating to float dtype
df['rating'] = df['rating'].astype(float)
```
